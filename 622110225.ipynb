# jaccard_similarity


```python
def jaccard_similarity(x,y):
    intersection_cardinality = len(set.intersection(*[set(x), set(y)]))
    union_cardinality = len(set.union(*[set(x), set(y)]))
    return intersection_cardinality/float(union_cardinality)
```

```python
def filterResult(distance):
  return distance > 0.7
generatorFunc(jaccard_similarity, filterResult, "jaccard_result")

```

# Needleman-Wunsch algorithm

```python
from itertools import product
from collections import deque


def needleman_wunsch(x, y):
    N, M = len(x), len(y)
    s = lambda a, b: int(a == b)
    DIAG = -1, -1
    LEFT = -1, 0
    UP = 0, -1

    # Create tables F and Ptr
    F = {}
    Ptr = {}

    F[-1, -1] = 0
    for i in range(N):
        F[i, -1] = -i
    for j in range(M):
        F[-1, j] = -j

    option_Ptr = DIAG, LEFT, UP
    for i, j in product(range(N), range(M)):
        option_F = (
            F[i - 1, j - 1] + s(x[i], y[j]),
            F[i - 1, j] - 1,
            F[i, j - 1] - 1,
        )
        F[i, j], Ptr[i, j] = max(zip(option_F, option_Ptr))

    # Work backwards from (N - 1, M - 1) to (0, 0)
    # to find the best alignment.
    alignment = deque()
    i, j = N - 1, M - 1
    while i >= 0 and j >= 0:
        direction = Ptr[i, j]
        if direction == DIAG:
            element = i, j
        elif direction == LEFT:
            element = i, None
        elif direction == UP:
            element = None, j
        alignment.appendleft(element)
        di, dj = direction
        i, j = i + di, j + dj
    while i >= 0:
        alignment.appendleft((i, None))
        i -= 1
    while j >= 0:
        alignment.appendleft((None, j))
        j -= 1


    score_gap = -1
    score_same = +1
    score_different = -1

    score = 0
    for i, j in alignment:
        if (i is None) or (j is None):
            score += score_gap
        elif x[i] == y[j]:
            score += score_same
        elif x[i] != y[j]:
            score += score_different

    return score
```

# Smith-waterman algorithm

```python
def smith_waterman(Q1, Q2, match_score=1, mismatch_score=-1, gap_score=-1):
    # Initialize the scoring matrix
    rows = len(Q1) + 1
    cols = len(Q2) + 1
    score_matrix = [[0 for j in range(cols)] for i in range(rows)]

    # Fill in the scoring matrix
    max_score = 0
    max_pos = 0, 0
    for i in range(1, rows):
        for j in range(1, cols):
            diag_score = score_matrix[i-1][j-1] + (match_score if Q1[i-1] == Q2[j-1] else mismatch_score)
            up_score = score_matrix[i-1][j] + gap_score
            left_score = score_matrix[i][j-1] + gap_score
            score = max(0, diag_score, up_score, left_score)
            score_matrix[i][j] = score
            if score > max_score:
                max_score = score
                max_pos = (i, j)

    # Initialize the aligned strings as the input strings
    Q1_aligned, Q2_aligned = Q1, Q2

    # Initialize the start position in the scoring matrix
    i, j = max_pos

    # Traceback to find the alignment
    if i > 0 and j > 0:
      while i > 0 and j > 0:
          score = score_matrix[i][j]
          diag_score = score_matrix[i-1][j-1]
          up_score = score_matrix[i-1][j]
          left_score = score_matrix[i][j-1]

          if score == 0:
              break
          elif score == diag_score + (match_score if Q1[i-1] == Q2[j-1] else mismatch_score):
              i -= 1
              j -= 1
          elif score == up_score + gap_score:
              i -= 1
              Q2_aligned = Q2_aligned[:j] + '-' + Q2_aligned[j:]
          elif score == left_score + gap_score:
              j -= 1
              Q1_aligned = Q1_aligned[:i] + '-' + Q1_aligned[i:]

      # Prepend the necessary gap symbols to get to (0,0)
      while i > 0:
          Q2_aligned = Q2_aligned[:j] + '-' + Q2_aligned[j:]
          i -= 1
      while j > 0:
          Q1_aligned = Q1_aligned[:i] + '-' + Q1_aligned[i:]
          j -= 1
      return max_score, Q1_aligned, Q2_aligned
    else:
      return 0, Q1, Q2
```

# Correlation TF-IDF

```python
import math
import pythainlp

def tf(word, document):
    # Count the frequency of the word in the document
    word_count = document.count(word)
    # Calculate the term frequency (TF)
    if word_count == 0:
        return 0
    else:
        return 1 + math.log10(word_count)

def idf(word, documents):
    # Count the number of documents containing the word
    document_count = sum(1 for doc in documents if word in doc)
    # Calculate the inverse document frequency (IDF)
    if document_count == 0:
        return 0
    else:
        return math.log10(len(documents) / document_count)

def tf_idf(word, document, documents):
    # Calculate the TF-IDF score for the word in the document
    return tf(word, document) * idf(word, documents)

def tokenize(text):
    # Tokenize the text into words using PyThaiNLP
    words = pythainlp.word_tokenize(text)
    # Remove any stop words from the list of words
    words = [word for word in words if word not in pythainlp.corpus.thai_stopwords()]
    return words

def calculate_tf_idf(documents):
    # Tokenize each document
    tokenized_documents = [tokenize(doc) for doc in documents]
    # Create a set of unique words in all the documents
    unique_words = set(word for doc in tokenized_documents for word in doc)
    # Calculate the TF-IDF score for each word in each document
    tf_idf_scores = []
    for i, doc in enumerate(tokenized_documents):
        scores = {word: tf_idf(word, doc, tokenized_documents) for word in unique_words}
        tf_idf_scores.append(scores)
    return tf_idf_scores
```

# Correlation Word2Vec

```python
import pythainlp
from gensim.models import Word2Vec

# Tokenize the input text into a list of sentences
def tokenize(text):
    sentences = pythainlp.sent_tokenize(text)
    return [pythainlp.word_tokenize(sentence) for sentence in sentences]

# Train the Word2Vec model on the input text
def train_word2vec(text, size=100, window=5, min_count=5, workers=4, sg=0):
    sentences = tokenize(text)
    model = Word2Vec(sentences, size=size, window=window, min_count=min_count, workers=workers, sg=sg)
    return model

# Find the most similar words to a given word in the Word2Vec model
def find_similar_words(word, model, topn=5):
    similar_words = model.wv.most_similar(word, topn=topn)
    return [word for word, similarity in similar_words]

# Example usage
text = "วันนี้ฟ้าสีฟ้า ใบไม้ร่วงลงมา พรุ่งนี้อากาศดี จะไปเที่ยวทะเล"
model = train_word2vec(text)
similar_words = find_similar_words("ฟ้า", model)
print(similar_words)
```
